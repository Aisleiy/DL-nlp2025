```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

# Example sentences
corpus = [
    'I love Python so deeply.',
    'Python is great for deep learning.',
    'I enjoy learning Machine Learning and Deep Learning.',
    'Deep Learning is an exciting in machine learning.',
    'Machine Learning is fascinating.',
    'Deep Learning for NLP class is so easy!'
]

# Step 1: Preprocessing and Tokenization
# This is handled by CountVectorizer in this case

# Step 2: Vocabulary Creation and Step 3: Vectorization
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

# Convert sparse matrix to DataFrame for better readability
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Display the DataFrame
print("Vocabulary:", vectorizer.get_feature_names_out())
print("Bag-of-Words Representation:\n", df)

# Visualization: Heatmap of the Bag-of-Words representation
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap="YlGnBu", cbar=False, xticklabels=True, yticklabels=[f'Doc {i+1}' for i in range(len(corpus))])
plt.title('Bag-of-Words Representation Heatmap')
plt.xlabel('Words')
plt.ylabel('Documents')
plt.xticks(rotation=90)
plt.show()
```
